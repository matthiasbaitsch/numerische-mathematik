# Stabilität von Lösungsverfahren

```julia
using CairoMakie
include("odesolve.jl")
```

## Funktion der DGL

```julia
F(t, y) = [
    y[2],
    100 * (1 - y[1]^2) * y[2] - y[1]
]
```

## Lösung für verschiedene Schrittweiten

Schrittweite $h = 0.001$

```julia
tEu, yEu = euler(     F, [0, 350], 0.001, [2, 0])
tRk, yRk = rungekutta(F, [0, 350], 0.001, [2, 0])
p = lines(tEu, yEu[:,1], label="Euler")
lines!(tRk, yRk[:,1], label="Runge-Kutta")
axislegend(position = :lb)
p
```

Schrittweite $h = 0.01$

```julia
tEu, yEu = euler(     F, [0, 350], 0.01, [2, 0])
tRk, yRk = rungekutta(F, [0, 350], 0.01, [2, 0])
p = lines(tEu, yEu[:,1], label="Euler")
lines!(tRk, yRk[:,1], label="Runge-Kutta")
axislegend(position = :lb)
p
```

Maximale Schrittweite

```julia
tEu, yEu = euler(     F, [0, 350], 0.004, [2, 0])
tRk, yRk = rungekutta(F, [0, 350], 0.00827, [2, 0])
p = scatterlines(tEu, yEu[:,1], label="Euler")
scatterlines!(tRk, yRk[:,1], label="Runge-Kutta")
axislegend(position = :lb)
p
```
