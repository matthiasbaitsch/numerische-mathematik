# Genauigkeit von Lösungsverfahren

```julia
using CairoMakie
include("odesolve.jl")
```

Funktion der DGL

```julia
F(t, y) = [y[2], t - y[1]]
```

Exakte Lösung

```julia
y(t) = [t + cos(t) - 2sin(t), 1 - 2cos(t) - sin(t)]
yex = stack(y.(0:0.1:20), dims=1)
lines(yex)
```

## Plotten

```julia
function plot_sol(sol)
    _, y1 = sol(F, [0, 20], 0.02, [1, -1])
    _, y2 = sol(F, [0, 20], 0.1, [1, -1])
    _, y3 = sol(F, [0, 20], 0.5, [1, -1])
    p = lines(yex, label="Exakt")
    scatterlines!(y1, label="h=0.02")
    scatterlines!(y2, label="h=0.1")
    scatterlines!(y3, label="h=0.5")
    axislegend()
    p
end
```

## Euler

```julia
plot_sol(euler)
```

## Halbschritt

```julia
plot_sol(halbschritt)
```

## Runge-Kutta

```julia
plot_sol(rungekutta)
```

## Bewertung

1. Euler: Die Funktionswerte weichen selbst für  signifikant von der exakten Lösung ab. 

1. Halbschritt: Etwas besser

1. Runge Kutta: Selbst für  erhält man bereits eine sehr gute Näherung.
