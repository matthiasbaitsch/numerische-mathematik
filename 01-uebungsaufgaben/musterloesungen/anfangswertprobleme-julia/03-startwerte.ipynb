# Stabilität von Lösungsverfahren

```julia
using CairoMakie
using LinearAlgebra
include("odesolve.jl")
```

## Funktion der DGL

```julia
sigma = 10
rho = 28
beta = 8 / 3

F(t, y) = [ 
        sigma * (y[2] - y[1]),
        y[1] * (rho - y[3]) - y[2],
        y[1] * y[2] - beta * y[3]
]
```

## Lösung für alpha = 5

```julia
h = 0.01
alpha = 5
t, ya = rungekutta(F, [0, 40], h, [10, 10, 10])
t, yb = rungekutta(F, [0, 40], h, [10, 10, 10 + alpha]);
```

Plot beider Kurven

```julia
p = lines(ya)
lines!(yb)
p
```

Vergleich

```julia
lines(norm.(eachrow(yb - ya)))
```

## Lösung für alpha = 1e-15

```julia
h = 0.01
alpha = 1e-15
t, ya = rungekutta(F, [0, 40], h, [10, 10, 10])
t, yb = rungekutta(F, [0, 40], h, [10, 10, 10 + alpha])

lines(norm.(eachrow(yb - ya)))
```
