# Genauigkeit von Lösungsverfahren

```python
import numpy as np
import matplotlib.pyplot as plt

from odesolve import *

%load_ext autoreload
%autoreload 2
```

Funktion der DGL

```python
def F(t, y):
    return np.array([y[1], t - y[0]])
```

Exakte Lösung

```python
t = np.linspace(0, 20, num=100)
yex1 = t + np.cos(t) - 2*np.sin(t)
yex2 = 1 - 2*np.cos(t) - np.sin(t)
plt.plot(yex1, yex2)
```

## Plotten

```python
def plot_sol(sol):
    _, y1 = sol(F, [0, 20], 0.02, [1, -1])
    _, y2 = sol(F, [0, 20], 0.1, [1, -1])
    _, y3 = sol(F, [0, 20], 0.5, [1, -1])
    plt.plot(yex1, yex2)
    plt.plot(y1[:,0],y1[:,1])
    plt.plot(y2[:,0],y2[:,1])
    plt.plot(y3[:,0],y3[:,1])
    plt.show()
```

## Euler

```python
plot_sol(euler)
```

## Halbschritt

```python
plot_sol(halbschritt)
```

## Runge-Kutta

```python
plot_sol(rungekutta)
```

## Bewertung

1. Euler: Die Funktionswerte weichen selbst für  signifikant von der exakten Lösung ab. 

1. Halbschritt: Etwas besser

1. Runge Kutta: Selbst für  erhält man bereits eine sehr gute Näherung.
