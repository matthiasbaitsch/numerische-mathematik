# Stabilität von Lösungsverfahren

```python
import numpy as np
import matplotlib.pyplot as plt

from odesolve import *

%load_ext autoreload
%autoreload 2
```

## Funktion der DGL

```python
def F(t, y):
    return np.array([
        y[1],
        100 * (1 - y[0]**2) * y[1] - y[0]
    ])
```

## Lösung für verschiedene Schrittweiten

Schrittweite $h = 0.001$

```python
tEu, yEu = euler(     F, np.array([0, 350]), 0.001, np.array([2, 0]))
tRk, yRk = rungekutta(F, np.array([0, 350]), 0.001, np.array([2, 0]))

plt.plot(tEu, yEu[:,0], label="Euler")
plt.plot(tRk, yRk[:,0], label="Runge-Kutta")
plt.legend(loc='lower left')
```

Schrittweite $h = 0.01$

```python
tEu, yEu = euler(     F, np.array([0, 350]), 0.01, np.array([2, 0]))
tRk, yRk = rungekutta(F, np.array([0, 350]), 0.01, np.array([2, 0]))

plt.plot(tEu, yEu[:,0], label="Euler")
plt.plot(tRk, yRk[:,0], label="Runge-Kutta")
plt.legend(loc='lower left')
```

Maximale Schrittweite

```python
tEu, yEu = euler(     F, [0, 350], 0.004, [2, 0])
tRk, yRk = rungekutta(F, [0, 350], 0.00827, [2, 0])

tEu, yEu = euler(     F, np.array([0, 350]), 0.004, np.array([2, 0]))
tRk, yRk = rungekutta(F, np.array([0, 350]), 0.008, np.array([2, 0]))

plt.plot(tEu, yEu[:,0], label="Euler")
plt.plot(tRk, yRk[:,0], label="Runge-Kutta")
plt.legend(loc='lower left')
```
